```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### 1. Read the dataset to the python environment

```python
cars_data =pd.read_csv("cars_data.csv")
cars_data
```

#### 2. Check for the null values present in the dataset.

```python
cars_data.isnull().sum()
```

#### 3. Plot a bar graph of male vs female buyers participated in the sales.

```python
plt.figure(figsize=(9,6))
cars_data["Buyer Gender"].value_counts().plot(kind="bar")
plt.title("Bar graph of male buyers v/s female buyers", fontsize=12)
plt.xlabel("Gender",fontsize=10)
plt.ylabel("Count", fontsize=10)
```

#### 4.Find the top 5 cars based on their sales price.

```python
cars_data1 = cars_data.sort_values(by=['Sale Price'], ascending=False)
cars_data1.head(5)
```

#### 5. Find the least 5 cars based on their Resell price.

```python
cars_data2 = cars_data.sort_values(by=['Resell Price'], ascending=True)
cars_data2.head(5)
```
